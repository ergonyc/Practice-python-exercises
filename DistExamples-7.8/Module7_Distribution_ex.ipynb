## 0 Setup 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_distributions(rand1,dist_name):
    #Plot a histogram for rand1.
    plt.hist(rand1, bins=20, color='c', density=True)
    # Add a vertical line at the mean.
    plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2) 
    # Add a vertical line at one standard deviation above/below the mean.
    plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)
    plt.axvline(rand1.mean() - rand1.std(), color='b', linestyle='dashed', linewidth=2) 
    # Print the histogram.
    plt.title(dist_name)
    plt.show()
```

## 1.1 Weibull (good psychometric curve)

\begin{equation}
p(x) = \frac{a}
{\lambda}(\frac{x}{\lambda})^{a-1}e^{-(x/\lambda)^a},
\end{equation}

for a = 1, this collapses ot an exponential.


```python
#weibull
npoints = 100
a = 5. # shape
s = np.random.weibull(a, npoints)

plot_distributions(s,'Weibull(5)')

```

## 1.2. Exponential
Its probability density function is

\begin{equation} f(x; \frac{1}{\beta}) = \frac{1}{\beta} \exp(-\frac{x}{\beta}), \end{equation}

for $x > 0$ and 0 elsewhere. $\beta$ is the scale parameter,
which is the inverse of the rate parameter $\lambda = 1/\beta$.
The rate parameter is an alternative, widely used parameterization
of the exponential distribution.


```python
#Exponential
s = np.random.exponential(2,npoints)

plot_distributions(s,'Exponential(2)')
    
    
```

## 1.3. Uniform


```python
# Uniform
s = np.random.uniform(0.0, 1.0, npoints)
plot_distributions(s,'Uniform(0,1)')
    
    
```

## 1.4. Gamma
The probability density for the Gamma distribution is
where k is the shape and $\theta$ the scale, and $\Gamma$ is the Gamma function.
\begin{equation}
p(x) = x^{k-1}\frac{e^{-x/\theta}}{\theta^k\Gamma(k)},
\end{equation}

```python
   
#gamma
shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)
s = np.random.gamma(shape, scale, npoints)
plot_distributions(s,'$\Gamma$(2.,2.)')

```

## 1.5. Chisquare
\begin{equation}
Q = \sum_{i=0}^{\mathtt{df}} X^2_i
is chi-square distributed, denoted
\end{equation}

\begin{equation}
Q \sim \chi^2_k.
\end{equation}

The probability density function of the chi-squared distribution is

\begin{equation}
p(x) = \frac{(1/2)^{k/2}}{\Gamma(k/2)}
x^{k/2 - 1} e^{-x/2},
\end{equation}

where $\Gamma$ is the gamma function,

\begin{equation}
\Gamma(x) = \int_0^{-\infty} t^{x - 1} e^{-t} dt.
\end{equation}

```python

#chisq
s=np.random.chisquare(4,npoints)
plot_distributions(s,'$X^2_4$')
```

## 1.6. Beta
\begin{equation}
f(x; a,b) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1}
(1 - x)^{\beta - 1},
\end{equation}
where the normalisation, $B$, is the beta function,

\begin{equation}
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1}
(1 - t)^{\beta - 1} dt.
\end{equation}

```python
#beta
s=np.random.beta(2,5,npoints)
plot_distributions(s,'$B(2,5)$' )
  
```
